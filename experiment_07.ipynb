# Segmentation Algorithm

```python
# Relevant Libraries
from skimage.filters import threshold_otsu
from skimage.morphology import remove_small_objects, remove_small_holes
import scipy.ndimage as ndimage
from skimage.morphology import convex_hull_image
from sklearn.cluster import KMeans
import numpy as np
```

```python
reference_img = mri_stacked[:,:,0]

# Normalize reference_img
reference_img = (reference_img - reference_img.min()) / (reference_img.max() - reference_img.min())

# Plot reference_img
plt.imshow(reference_img, cmap='gray')
plt.axis('off')
plt.title('Reference Image')
plt.show()

temporal_masks = {}

# Apply an outsu algorithm to the reference image
threshold = threshold_otsu(reference_img)
binary = reference_img > threshold - 0.02

# Remove the smaller artificts (holes and objects)
cleaned = remove_small_holes(binary)
cleaned = remove_small_objects(cleaned, min_size=1000)

# Plot Cleaned Image
plt.imshow(cleaned, cmap='gray')
plt.axis('off')
plt.title('Binarized Cleaned Image')

# Apply two-pass algorithm to the cleaned image to have preliminary labels
labels, num_labels = ndimage.label(cleaned)

# NOTE: I got 3 labels, but I save only the first label
# Save the temporal mask label 1
temporal_masks["1"] = (labels == 1).astype(int)

# Plot temporal masks
utils.plot_masks(temporal_masks)
```

```python
# Invert the cleaned image
inverted = ~cleaned

# Got the others labels from the inverted image
# using the same two-pass algorithm
inverted_labels, num_labels = ndimage.label(inverted)

# Only consider the two first labels and all the other to zero
inverted_labels[inverted_labels > 2] = 0

# Save temporal masks label 0 and 2
temporal_masks["0"] = (inverted_labels == 1).astype(int)
temporal_masks["2"] = (inverted_labels == 2).astype(int)

# Got the tmp internal region by considering all that is
# not part of the mask 0, 1, 2
tmp_mask = temporal_masks["0"] + temporal_masks["1"] + temporal_masks["2"]
temporal_masks["tmp_internal_region"] = (tmp_mask == 0).astype(int)

# Plot temporal masks
utils.plot_masks(temporal_masks, rows=2, cols=2)

# Plot the temporal_mask in a unique img


keys = ["0", "1", "2", "tmp_internal_region"]

utils.plot_temporal_masks_by_keys(temporal_masks, keys, title = "Four Masks Labelling")
```

```python
from skimage.feature import canny

# Apply temporal mask 2 to the reference image
masked = reference_img * temporal_masks["2"]

# Apply k-means to the masked image
labels, order = utils.kmeans_segmentation(masked, n_clusters=3)

temporal_masks["tmp_mask2"] = (labels == order[1]).astype(int)

# Plot temporal masks
utils.plot_masks(temporal_masks, rows=2, cols=3)
```

```python
# Apply the convex hull to the tmo_internal_region
convex_hull = convex_hull_image(temporal_masks["tmp_internal_region"])

# Plot the binary image
plt.imshow(convex_hull, cmap='gray')
plt.title('Convex Hull Internal Mask')
plt.axis('off')
plt.show()

# Save the convex hull mask in the temporal masks
temporal_masks["convex_hull"] = convex_hull.astype(int)

# Apply the convex hull to the mask1
masked_convex_hull = reference_img * convex_hull

# Plot the masked convex hull
plt.imshow(masked_convex_hull, cmap='gray')
plt.show()
```

```python
keys = ["0", "1", "2", "convex_hull"]

utils.plot_temporal_masks_by_keys(temporal_masks, keys, title = "Internal Mask Refinement")
```

```python
from skimage.feature import canny

# Get the edges from masked_convex_hull
edges = canny(masked_convex_hull, sigma=2)

# Dilate the edges
edges = ndimage.binary_dilation(edges)

# Fill holes in the edges
edges = ndimage.binary_fill_holes(edges)

# remove small objects
edges = remove_small_objects(edges, min_size=100)

# Plot the edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Edges-based Mask')
plt.show()
```

```python
# Apply edges to the masked_convex_hull
masked_internal_region = reference_img * edges

# Apply k-means clustering to the partial_reference image
internal_labels, order = utils.kmeans_segmentation(masked_internal_region, n_clusters=4)

# Use the order to fit the mask 3, 4, 5
# NOTE: I don't use order 0 because it corresponds to the external region
temporal_masks["3"] = (internal_labels == order[1]).astype(int)
temporal_masks["4"] = (internal_labels == order[2]).astype(int)
temporal_masks["5"] = (internal_labels == order[3]).astype(int)

# Plot the temporal masks
utils.plot_masks(temporal_masks, rows=3, cols=4)
```

```python
# NOTE: There are certain difference between the convex hull 
# and the internal region, which consequently affect the
# k-means clustering, and mask 2. Then, I need to refine
# the mask 2

# Refine the mask 2 by considering all the others masks
tmp_mask = ( temporal_masks["0"] + temporal_masks["1"] +
            temporal_masks["3"] + temporal_masks["4"] +
            temporal_masks["5"] )
temporal_masks["2"] = (tmp_mask == 0).astype(int)

# Plot the temporal masks
utils.plot_masks(temporal_masks, rows=3, cols=4)
```

```python
keys = ["0", "1", "2", "3", "4", "5"]

utils.plot_temporal_masks_by_keys(temporal_masks, keys, title = "Internal Mask Labelling")
```

```python
# Refine temporal mask 1 and 2 using the temp_mask_2

# Sum the masks 1, 2
tmp_mask = ((temporal_masks["1"] + temporal_masks["2"] + temporal_masks["tmp_mask2"]) > 0).astype(int)

# refinement
refinement = tmp_mask - temporal_masks["tmp_mask2"]

# # Remove less connected objects
# refinement = remove_small_objects(refinement, min_size=100)

# Plot refinement
plt.imshow(refinement, cmap='gray')
plt.axis('off')
plt.title('Refinement')
plt.show()

# Save refinement
temporal_masks["1"] = refinement
temporal_masks["2"] = temporal_masks["tmp_mask2"]
```
